Exercise 26

````python
from google.colab import files
uploaded = files.upload()
def file_stats(filename: str) -> None:
    """Print total lines, words, and characters in the given text file."""
    num_lines = num_words = num_chars = 0

    with open(filename, "r", encoding="utf-8") as f:
        for line in f:
            num_lines += 1
            num_words += len(line.split())
            num_chars += len(line)            # includes newline chars

    print(f"Lines      : {num_lines}")
    print(f"Words      : {num_words}")
    print(f"Characters : {num_chars}")

if __name__ == "__main__":
    file_stats("essay.txt")
````

````python
from google.colab import files
uploaded = files.upload()
from collections import Counter
import string

def count_letter_frequency(filename):
    with open(filename, "r", encoding="utf-8") as file:
        text = file.read().lower()  # convert to lowercase for case-insensitive counting

    # Filter only alphabetic letters
    letters_only = [char for char in text if char in string.ascii_lowercase]

    # Count letter frequencies
    frequency = Counter(letters_only)

    # Print the frequencies sorted by letter
    for letter in sorted(frequency):
        print(f"{letter}: {frequency[letter]}")

# Example usage
count_letter_frequency("message.txt")
````

````python
from google.colab import files
uploaded = files.upload()
from collections import Counter
import string

def most_frequent_word(filename):
    with open(filename, "r", encoding="utf-8") as file:
        text = file.read().lower()  # Convert to lowercase for case-insensitive counting

    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    clean_text = text.translate(translator)

    words = clean_text.split()

    if not words:
        print("No words found in the file.")
        return

    word_counts = Counter(words)
    most_common_word, frequency = word_counts.most_common(1)[0]

    print(f"The most frequent word is '{most_common_word}' with {frequency} occurrences.")

# Example usage
most_frequent_word("book.txt")
````

````python
from google.colab import files
uploaded = files.upload()
import os

def is_file_writable(filename):
    # Check if file exists first
    if not os.path.exists(filename):
        print(f"File '{filename}' does not exist.")
        return False

    # Check write permission
    if os.access(filename, os.W_OK):
        print(f"File '{filename}' is writable.")
        return True
    else:
        print(f"File '{filename}' is not writable.")
        return False

# Example usage
is_file_writable("students.txt")
````

````python
from google.colab import files
uploaded = files.upload()
import string

def count_unique_words(filename):
    with open(filename, "r", encoding="utf-8") as file:
        text = file.read().lower()  # lowercase for case-insensitive counting

    # Remove punctuation
    translator = str.maketrans('', '', string.punctuation)
    clean_text = text.translate(translator)

    words = clean_text.split()
    unique_words = set(words)

    return len(unique_words)

# Example usage
unique_count = count_unique_words("destination.txt")
print(f"Number of unique words: {unique_count}")
````

````python
from google.colab import files
uploaded = files.upload()
from collections import Counter
import string

def most_common_letter(filename):
    with open(filename, "r", encoding="utf-8") as file:
        text = file.read().lower()

    # Filter only letters a-z
    letters = [char for char in text if char in string.ascii_lowercase]

    if not letters:
        return None  # No letters found

    letter_counts = Counter(letters)
    most_common = letter_counts.most_common(1)[0]  # (letter, count)

    return most_common

# Example usage
result = most_common_letter("example.txt")
if result:
    print(f"The most common letter is '{result[0]}' with {result[1]} occurrences.")
else:
    print("No letters found in the file.")
````
